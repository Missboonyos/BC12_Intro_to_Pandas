**Basic Pandas**

```python
import pandas as pd
```

```python
# Create dataframe from scratch
raw_data = {
    "name": ["Jan", "Febby", "Marchy", "Abril", "May"],
    "age" : [20, 21, 19, 18, 23],
    "gender": ["M", "F", "M", "F", "M"]
}

df = pd.DataFrame(raw_data) # to reuse the result


```

```python
df
```

```python
# Add new column
df['city'] = ["Prague", "London", "Frankfurt", "Berlin", "Bangkok"]
```

```python
# Check Shape
df.shape
```

```python
df
```

```python
# drop city
df = df.drop("city", axis=1)
```

```python
df
```

```python
# remove index = 2
df = df.drop(2, axis=0)
```

```python
# create a new series
s1 = pd.Series(['Mary', 20, 'F'], index=['nickname', 'age', 'sex'])
print(s1)
print(type(s1))
```

```python
# append s1 to df # Append using concat (modern Pandas)
df = pd.concat([df, s1.to_frame().T], ignore_index=True)
df
```

```python
df['age'] = df['age'].astype(int)
```

```python
df['age']
```

```python
# create a new column
s2 = pd.Series(['London', 'New York', 'Berlin', 'Madrid', 'Surin'])
df['city'] = s2
df
```

```python
# write csv file
df.to_csv('new_df.csv')
```
